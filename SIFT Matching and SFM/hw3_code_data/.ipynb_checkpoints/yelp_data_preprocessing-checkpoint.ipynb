```python
import pandas as pd
import ast
```

```python
df=pd.read_csv("./yelp_responses.csv")
rows=[]
for index, row in df.iterrows():
    if not pd.isnull(row['businesses']):
        output = ast.literal_eval(row['businesses'])
        region_latitue=row['region.center.latitude']
        region_longitude=row['region.center.longitude']
        curr_list=list(output)
        for i in range(len(curr_list)):
            curr_list[i]["region_latitue"]=region_latitue
            curr_list[i]["region_longitude"]=region_longitude
            values=list(curr_list[i].values())
            rows.append(values)
finalDataFrame=pd.DataFrame(rows)
```

```python
finalDataFrame.columns = ['id','alias','name','image_url','is_closed','url','review_count', 'categories', 'rating',
                          'coordinates', 'transactions','price','location','phone','display_phone','distance', 'region_latitue', 'region_longitude']
```

```python
finalDataFrame.insert(9, 'latitude', 0)
finalDataFrame.insert(10, 'longitude',0)
```

```python
finalDataFrame['coordinates']=finalDataFrame['coordinates'].astype('str')
finalDataFrame[['latitude','longitude']]=finalDataFrame['coordinates'].str.extractall('(\d*\.\d+|\d+.*)').unstack()
```

```python
finalDataFrame.drop(columns=['coordinates'])
```

```python
finalDataFrame.to_csv(r"./yelp_responses_preprocessed.csv")
```
